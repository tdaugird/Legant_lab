### This is some initial script for looking at intensities of PPARg around the periphery of the CEBPb hubs. This will make mean intensites for categories. it will also look at intensities in regions as the rings get larger and larger around the periphery of hubs

```python
'''generated lists that contain the names of related categories'''

cat_name_dict = {'area_heteroc' : [],
                'dapi_heteroc' : [],
                'cebp_heteroc' : [],
                'ppar_heteroc' : [],
                'area_blob' : [],
                'dapi_blob' : [],
                'cebp_blob' : [],
                'ppar_blob' : []}

'''iterate over all of the categories'''
for column in list(df):
    for key in cat_name_dict:
        key_ = str(key)
        column_ = str(column)
        if key_[:4] == column_[:4] and key_[-4:] == column_[-4:] and '_n_' not in column_:
            cat_name_dict[key].append(column)
```

```python
'''drops the toal area and reverses the order of the m# entries in order to make things go in increasing order'''
for key in cat_name_dict:
    value = cat_name_dict[key]
    value = value[1:]
    first4 = value[:4]
    first4 = first4[::-1]
    cat_name_dict[key] = cat_name_dict[key][1:]
    cat_name_dict[key][:4] = first4
    
```
