### This is notebook to look at the .csv that determines intensity of different fluorescent chanels in and around the heterochromatin spaces and cebp hubs.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import seaborn as sns
```

```python
df_orig = pd.read_csv('running_day3_dilations.csv')
```

```python
''' this is some housekeeping to for plots etc. '''

df_orig['rand'] = np.random.randint(0, 19, len(df_orig))
ppar_palette = sns.color_palette('PuRd', 20)
cebp_palette = sns.color_palette('YlGn', 20)
dapi_palette = sns.color_palette('Blues', 20)
area_palette = sns.color_palette('Greys', 20)

df = df_orig.copy()
```

### First thing that I am going to do is figure out the mean intensity in each of the areas

```python
'''mean intensities inside the dapi mask'''
df_orig['mean_dapi_nucleus'] = np.divide(df.dapi_nucleus, df.area_nucleus)
df_orig['mean_cebp_nucleus'] = np.divide(df.cebp_nucleus, df.area_nucleus)
df_orig['mean_ppar_nucleus'] = np.divide(df.ppar_nucleus, df.area_nucleus)

df_orig['mean_dapi_heteroc'] = np.divide(df.dapi_heteroc, df.area_heteroc)
df_orig['mean_cebp_heteroc'] = np.divide(df.cebp_heteroc, df.area_heteroc)
df_orig['mean_ppar_heteroc'] = np.divide(df.ppar_heteroc, df.area_heteroc)

'''mean intensities outside of the dapi mask and in the nucleus mask'''
df_orig['mean_dapi_n_heteroc'] = np.divide(df.dapi_n_heteroc, df.area_n_heteroc)
df_orig['mean_cebp_n_heteroc'] = np.divide(df.cebp_n_heteroc, df.area_n_heteroc)
df_orig['mean_ppar_n_heteroc'] = np.divide(df.ppar_n_heteroc, df.area_n_heteroc)

'''maen intensites inside the cebp hubs'''
df_orig['mean_dapi_blob'] = np.divide(df.dapi_blob, df.area_blob)
df_orig['mean_cebp_blob'] = np.divide(df.cebp_blob, df.area_blob)
df_orig['mean_ppar_blob'] = np.divide(df.ppar_blob, df.area_blob)

'''mean intensities outside of the cebp hub and in the nucleus mask'''
df_orig['mean_dapi_n_blob'] = np.divide(df.dapi_n_blob, df.area_n_blob)
df_orig['mean_cebp_n_blob'] = np.divide(df.cebp_n_blob, df.area_n_blob)
df_orig['mean_ppar_n_blob'] = np.divide(df.ppar_n_blob, df.area_n_blob)
```

```python
'''determine mean intensity in the eroded heterochromatine regions'''
'''determine the mean intensity for each area around the heterochromatin'''

area_heteroc_name_lst = []
dapi_heteroc_name_lst = []
cebp_heteroc_name_lst = []
ppar_heteroc_name_lst = []

for col in list(df_orig): #iterate over all of the columns in the df and pull out the ones that are the area of rings around heterochromatin
    if 'heteroc' in str(col):
        if 'area_m' in str(col) or 'area_p' in str(col):
            area_heteroc_name_lst.append(col)
        if 'dapi_m' in str(col) or 'dapi_p' in str(col):
            dapi_heteroc_name_lst.append(col)
        if 'cebp_m' in str(col) or 'cebp_p' in str(col):
            cebp_heteroc_name_lst.append(col)
        if 'ppar_m' in str(col) or 'ppar_p' in str(col):
            ppar_heteroc_name_lst.append(col)
for area in range(len(area_heteroc_name_lst)): #make categoiries that are the mean intensity in each of the designated reings around the heteroc
    df_orig['mean_'+dapi_heteroc_name_lst[area]] = np.divide(df_orig[dapi_heteroc_name_lst[area]], df_orig[area_heteroc_name_lst[area]])
    df_orig['mean_'+cebp_heteroc_name_lst[area]] = np.divide(df_orig[dapi_heteroc_name_lst[area]], df_orig[area_heteroc_name_lst[area]])
    df_orig['mean_'+ppar_heteroc_name_lst[area]] = np.divide(df_orig[dapi_heteroc_name_lst[area]], df_orig[area_heteroc_name_lst[area]])

    
'''determine the mean intensity in each of the cebp blob areas''' 
area_blob_name_lst = []
dapi_blob_name_lst = []
cebp_blob_name_lst = []
ppar_blob_name_lst = []

for col in list(df_orig): #iterarte over teh columns of he df and pull out hte ones that are increasingly larger rings around the cebpb blobs
    if 'blob' in str(col):
        if 'area_m' in str(col) or 'area_p' in str(col):
            area_blob_name_lst.append(col)
        if 'dapi_m' in str(col) or 'dapi_p' in str(col):
            dapi_blob_name_lst.append(col)
        if 'cebp_m' in str(col) or 'cebp_p' in str(col):
            cebp_blob_name_lst.append(col)
        if 'ppar_m' in str(col) or 'ppar_p' in str(col):
            ppar_blob_name_lst.append(col)
for area in range(len(area_blob_name_lst)): #determine the mean intensity for each fluor. channel in the designated areas
    df_orig['mean_'+dapi_blob_name_lst[area]] = np.divide(df_orig[dapi_blob_name_lst[area]], df_orig[area_blob_name_lst[area]])
    df_orig['mean_'+cebp_blob_name_lst[area]] = np.divide(df_orig[dapi_blob_name_lst[area]], df_orig[area_blob_name_lst[area]])
    df_orig['mean_'+ppar_blob_name_lst[area]] = np.divide(df_orig[dapi_blob_name_lst[area]], df_orig[area_blob_name_lst[area]])
```

```python
'''reverse the first few elements in these lists. It will make the integrated areas and intensities script below work a little better'''
names_lst_o_lst = [area_heteroc_name_lst, dapi_heteroc_name_lst, cebp_heteroc_name_lst, ppar_heteroc_name_lst,
                  area_blob_name_lst, dapi_blob_name_lst, cebp_blob_name_lst, ppar_blob_name_lst]

for lst in names_lst_o_lst:
    lst_beg = lst[:4]
    lst_rev = lst_beg[::-1]
    lst[:4] = lst_rev
```

```python
for i in range(len(area_heteroc_name_lst)):
    print(area_heteroc_name_lst[i], dapi_heteroc_name_lst[i], cebp_heteroc_name_lst[i], ppar_heteroc_name_lst[i])
    #print(area_blob_name_lst[i], dapi_blob_name_lst[i], cebp_blob_name_lst[i], ppar_blob_name_lst[i])
```

```python
'''make parameters that are the intesities over integrated area arond the heterochromatin . This will make a new column for integrated areas and intensities.
It then uses this newly created column for creating the next integrated area'''

area_lst = np.copy(area_heteroc_name_lst)
dapi_lst = np.copy(dapi_heteroc_name_lst)
cebp_lst = np.copy(cebp_heteroc_name_lst)
ppar_lst = np.copy(ppar_heteroc_name_lst)


'''set up the first elements for generating parameters for mean intensities over integrated area'''
area_array1 = df_orig[area_heteroc_name_lst[0]]
dapi_array1 = df_orig[dapi_heteroc_name_lst[0]]
cebp_array1 = df_orig[cebp_heteroc_name_lst[0]]
ppar_array1 = df_orig[ppar_heteroc_name_lst[0]]

for i in range(1, len(area_lst)):
    area_array2 = df_orig[area_lst[i]] #make array of chosen element
    df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]] = area_array2 + area_array1 #make a new category that is integrated area
    df_orig['normed_'+area_lst[i][:5] + 'integ_' + area_lst[i][5:]] = np.divide(df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]],
                                                                                  df_orig.area_nucleus)
    area_array1 = df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]]
    
    
    dapi_array2 = df_orig[dapi_lst[i]]
    df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = dapi_array2 + dapi_array1
    df_orig['mean_'+dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = (dapi_array2 + dapi_array1) / (area_array2 + area_array1)
    df_orig['normed_'+dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = np.divide(df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]],
                                                                                  df_orig.mean_dapi_nucleus)
    dapi_array1 = df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]]
    
                                                                                  
    cebp_array2 = df_orig[cebp_lst[i]]
    df_orig[cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = cebp_array2 + cebp_array1
    df_orig['mean_'+cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = (cebp_array2 + cebp_array1) / (area_array2 + area_array1)
    df_orig['normed_'+cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = np.divide(df_orig[cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]],
                                                                                 df_orig.mean_cebp_nucleus)
    cebp_array1 = df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]]
    
    ppar_array2 = df_orig[ppar_lst[i]]
    df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = ppar_array2 + ppar_array1
    df_orig['mean_' + ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = (ppar_array2 + ppar_array1) / (area_array2 + area_array1)
    df_orig['normed_'+ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = np.divide(df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]],
                                                                                df_orig.mean_ppar_nucleus)
    ppar_array1 = df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]]
    
```

```python
'''this does the same thing as the code above it just looks at area around the CEBPb blobs rather than the heterochromating.
make parameters that are the intesities over integrated area arond the heterochromatin . This will make a new column for integrated areas and intensities.
It then uses this newly created column for creating the next integrated area'''

area_lst = np.copy(area_blob_name_lst)
dapi_lst = np.copy(dapi_blob_name_lst)
cebp_lst = np.copy(cebp_blob_name_lst)
ppar_lst = np.copy(ppar_blob_name_lst)


'''set up the first elements for generating parameters for mean intensities over integrated area'''
area_array1 = df_orig[area_blob_name_lst[0]]
dapi_array1 = df_orig[dapi_blob_name_lst[0]]
cebp_array1 = df_orig[cebp_blob_name_lst[0]]
ppar_array1 = df_orig[ppar_blob_name_lst[0]]

for i in range(1, len(area_lst)):
    area_array2 = df_orig[area_lst[i]] #make array of chosen element
    df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]] = area_array2 + area_array1 #make a new category that is integrated area
    df_orig['normed_'+area_lst[i][:5] + 'integ_' + area_lst[i][5:]] = np.divide(df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]],
                                                                                  df_orig.area_nucleus)
    area_array1 = df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]]
    
    
    dapi_array2 = df_orig[dapi_lst[i]]
    df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = dapi_array2 + dapi_array1
    df_orig['normed_'+dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = np.divide(df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]],
                                                                                  df_orig.mean_dapi_nucleus)
    dapi_array1 = df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]]
    
                                                                                  
    cebp_array2 = df_orig[cebp_lst[i]]
    df_orig[cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = cebp_array2 + cebp_array1
    df_orig['normed_'+cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = np.divide(df_orig[cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]],
                                                                                 df_orig.mean_cebp_nucleus)
    cebp_array1 = df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]]
    
    ppar_array2 = df_orig[ppar_lst[i]]
    df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = ppar_array2 + ppar_array1
    df_orig['normed_'+ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = np.divide(df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]],
                                                                                df_orig.mean_ppar_nucleus)
    ppar_array1 = df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]]
```
