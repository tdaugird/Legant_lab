```python
'''mean intensities inside the dapi mask'''
df_orig['mean_dapi_nucleus'] = np.divide(df_orig.dapi_nucleus, df_orig.area_nucleus)
df_orig['mean_cebp_nucleus'] = np.divide(df_orig.cebp_nucleus, df_orig.area_nucleus)
df_orig['mean_ppar_nucleus'] = np.divide(df_orig.ppar_nucleus, df_orig.area_nucleus)

df_orig['mean_dapi_heteroc'] = np.divide(df_orig.dapi_heteroc, df_orig.area_heteroc)
df_orig['mean_cebp_heteroc'] = np.divide(df_orig.cebp_heteroc, df_orig.area_heteroc)
df_orig['mean_ppar_heteroc'] = np.divide(df_orig.ppar_heteroc, df_orig.area_heteroc)

'''mean intensities outside of the dapi mask and in the nucleus mask'''
df_orig['mean_dapi_n_heteroc'] = np.divide(df_orig.dapi_n_heteroc, df_orig.area_n_heteroc)
df_orig['mean_cebp_n_heteroc'] = np.divide(df_orig.cebp_n_heteroc, df_orig.area_n_heteroc)
df_orig['mean_ppar_n_heteroc'] = np.divide(df_orig.ppar_n_heteroc, df_orig.area_n_heteroc)

'''maen intensites inside the cebp hubs'''
df_orig['mean_dapi_blob'] = np.divide(df_orig.dapi_blob, df_orig.area_blob)
df_orig['mean_cebp_blob'] = np.divide(df_orig.cebp_blob, df_orig.area_blob)
df_orig['mean_ppar_blob'] = np.divide(df_orig.ppar_blob, df_orig.area_blob)

'''mean intensities outside of the cebp hub and in the nucleus mask'''
df_orig['mean_dapi_n_blob'] = np.divide(df_orig.dapi_n_blob, df_orig.area_n_blob)
df_orig['mean_cebp_n_blob'] = np.divide(df_orig.cebp_n_blob, df_orig.area_n_blob)
df_orig['mean_ppar_n_blob'] = np.divide(df_orig.ppar_n_blob, df_orig.area_n_blob)
```

```python
'''determine mean intensity in the eroded heterochromatine regions'''
'''determine the mean intensity for each area around the heterochromatin'''

area_heteroc_name_lst = []
dapi_heteroc_name_lst = []
cebp_heteroc_name_lst = []
ppar_heteroc_name_lst = []

for col in list(df_orig): #iterate over all of the columns in the df and pull out the ones that are the area of rings around heterochromatin
    if 'heteroc' in str(col):
        if 'area_m' in str(col) or 'area_p' in str(col):
            area_heteroc_name_lst.append(col)
        if 'dapi_m' in str(col) or 'dapi_p' in str(col):
            dapi_heteroc_name_lst.append(col)
        if 'cebp_m' in str(col) or 'cebp_p' in str(col):
            cebp_heteroc_name_lst.append(col)
        if 'ppar_m' in str(col) or 'ppar_p' in str(col):
            ppar_heteroc_name_lst.append(col)
for area in range(len(area_heteroc_name_lst)): #make categoiries that are the mean intensity in each of the designated reings around the heteroc
    df_orig['mean_'+dapi_heteroc_name_lst[area]] = np.divide(df_orig[dapi_heteroc_name_lst[area]], df_orig[area_heteroc_name_lst[area]])
    df_orig['mean_'+cebp_heteroc_name_lst[area]] = np.divide(df_orig[dapi_heteroc_name_lst[area]], df_orig[area_heteroc_name_lst[area]])
    df_orig['mean_'+ppar_heteroc_name_lst[area]] = np.divide(df_orig[dapi_heteroc_name_lst[area]], df_orig[area_heteroc_name_lst[area]])

    
'''determine the mean intensity in each of the cebp blob areas''' 
area_blob_name_lst = []
dapi_blob_name_lst = []
cebp_blob_name_lst = []
ppar_blob_name_lst = []

for col in list(df_orig): #iterarte over teh columns of he df and pull out hte ones that are increasingly larger rings around the cebpb blobs
    if 'blob' in str(col):
        if 'area_m' in str(col) or 'area_p' in str(col):
            area_blob_name_lst.append(col)
        if 'dapi_m' in str(col) or 'dapi_p' in str(col):
            dapi_blob_name_lst.append(col)
        if 'cebp_m' in str(col) or 'cebp_p' in str(col):
            cebp_blob_name_lst.append(col)
        if 'ppar_m' in str(col) or 'ppar_p' in str(col):
            ppar_blob_name_lst.append(col)
for area in range(len(area_blob_name_lst)): #determine the mean intensity for each fluor. channel in the designated areas
    df_orig['mean_'+dapi_blob_name_lst[area]] = np.divide(df_orig[dapi_blob_name_lst[area]], df_orig[area_blob_name_lst[area]])
    df_orig['mean_'+cebp_blob_name_lst[area]] = np.divide(df_orig[dapi_blob_name_lst[area]], df_orig[area_blob_name_lst[area]])
    df_orig['mean_'+ppar_blob_name_lst[area]] = np.divide(df_orig[dapi_blob_name_lst[area]], df_orig[area_blob_name_lst[area]])
```

```python
'''reverse the first few elements in these lists. It will make the integrated areas and intensities script below work a little better'''
names_lst_o_lst = [area_heteroc_name_lst, dapi_heteroc_name_lst, cebp_heteroc_name_lst, ppar_heteroc_name_lst,
                  area_blob_name_lst, dapi_blob_name_lst, cebp_blob_name_lst, ppar_blob_name_lst]

for lst in names_lst_o_lst:
    lst_beg = lst[:4]
    lst_rev = lst_beg[::-1]
    lst[:4] = lst_rev
```

```python
for i in range(len(area_heteroc_name_lst) -20):
    print(area_heteroc_name_lst[i], dapi_heteroc_name_lst[i], cebp_heteroc_name_lst[i], ppar_heteroc_name_lst[i])
    #print(area_blob_name_lst[i], dapi_blob_name_lst[i], cebp_blob_name_lst[i], ppar_blob_name_lst[i])
```

```python
'''make parameters that are the intesities over integrated area arond the heterochromatin . This will make a new column for integrated areas and intensities.
It then uses this newly created column for creating the next integrated area'''

area_lst = np.copy(area_heteroc_name_lst)
dapi_lst = np.copy(dapi_heteroc_name_lst)
cebp_lst = np.copy(cebp_heteroc_name_lst)
ppar_lst = np.copy(ppar_heteroc_name_lst)


'''set up the first elements for generating parameters for mean intensities over integrated area'''
area_array1 = df_orig[area_lst[0]]
dapi_array1 = df_orig[dapi_lst[0]]
cebp_array1 = df_orig[cebp_lst[0]]
ppar_array1 = df_orig[ppar_lst[0]]

for i in range(1, len(area_lst)):
    area_array2 = df_orig[area_lst[i]] #make array of chosen element
    integrated_area = area_array1 + area_array2
    df_orig[area_lst[i][:5] + 'integ_' + area_lst[i][5:]] = integrated_area #make a new category that is integrated area
    df_orig['normed_'+area_lst[i][:5] + 'integ_' + area_lst[i][5:]] = integrated_area / df_orig.area_nucleus
    area_array1 = integrated_area
    
    
    dapi_array2 = df_orig[dapi_lst[i]] #pull out dapi intensity in the next largest region
    integrated_dapi = dapi_array1 + dapi_array2 #add the intensity in this region to the running lst
    df_orig[dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = integrated_dapi #maek a new category that is intensity in increasingly large region aroujnd heterochromating
    df_orig['mean_'+dapi_lst[i][:5] + 'integ_' + dapi_lst[i][5:]] = integrated_dapi / integrated_area #get mean intensity in this region
    relative_density_dapi = (integrated_dapi / integrated_area) / df_orig.mean_dapi_nucleus # divide mean intensity in this region by mean intensity across nuclues
    df_orig['relative_density_'+dapi_lst[i]] = relative_density_dapi #make new cat that includes this parameter
    dapi_array1 = integrated_dapi #make the ####_array1 the integrated intensity of this region
    
                                                                                  
    cebp_array2 = df_orig[cebp_lst[i]] #pull out cebp intensity in the next largest region
    integrated_cebp = cebp_array1 + cebp_array2 #add the intensity in this region to the running lst
    df_orig[cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = integrated_cebp #maek a new category that is intensity in increasingly large region aroujnd heterochromating
    df_orig['mean_'+cebp_lst[i][:5] + 'integ_' + cebp_lst[i][5:]] = integrated_cebp / integrated_area #get mean intensity in this region
    relative_density_cebp = (integrated_cebp / integrated_area) / df_orig.mean_cebp_nucleus # divide mean intensity in this region by mean intensity across nuclues
    df_orig['relative_density_'+cebp_lst[i]] = relative_density_cebp #make new cat that includes this parameter
    cebp_array1 = integrated_cebp #make the ####_array1 the integrated intensity of this region
    
    ppar_array2 = df_orig[ppar_lst[i]] #pull out ppar intensity in the next largest region
    integrated_ppar = ppar_array1 + ppar_array2 #add the intensity in this region to the running lst
    df_orig[ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = integrated_ppar #maek a new category that is intensity in increasingly large region aroujnd heterochromating
    df_orig['mean_'+ppar_lst[i][:5] + 'integ_' + ppar_lst[i][5:]] = integrated_ppar / integrated_area #get mean intensity in this region
    relative_density_ppar = (integrated_ppar / integrated_area) / df_orig.mean_ppar_nucleus # divide mean intensity in this region by mean intensity across nuclues
    df_orig['relative_density_'+ppar_lst[i]] = relative_density_ppar #make new cat that includes this parameter
    ppar_array1 = integrated_ppar #make the ####_array1 the integrated intensity of this region
    
```

```python
relative_density_cat_dict = {'dapi_heteroc' : [],
                            'cebp_heteroc' : [],
                            'ppar_heteroc' : [],
                            'dapi_blob' : [],
                            'cebp_blob' : [],
                            'ppar_blob' : []
                            }
for cat in list(df_orig):
    if 'relative_density' in str(cat):
        for key in relative_density_cat_dict:
            key_ = str(key)
            cat_ = str(cat)
            if cat_[17:21] in key_ and cat_[-4:] == key_[-4:]: # and cat_[-4:] in key_:
                relative_density_cat_dict[key].append(cat_)
```

```python
'''look at relative density of dapi relative to the heterochromatin regions'''
category = 'dapi_heteroc' #the key in relative_density_cat_dict that corresponds with the list of categories that I am interested in

parse_lst = relative_density_cat_dict[category] #make a list of categories to choose from entire df

dfs_lst_ = [] #running list of dfs that are generated in the following script.

for i in np.unique(df_orig.segm_img_name): #iterate over each of the image entries in the original df
    df_ = df_orig.loc[df_orig.segm_img_name == i] #choose a single image 
    
    img_name_ = df_.segm_img_name #pull out the image name for bookkeeping category
    #rand = int(df_.rand) # this is book keeping that will make data visualizaiton cleaner

    df_ = df_[parse_lst] #pares df_ further to only included the integrated intensity entries
    
    relative_density_lst = np.array(df_.iloc[0, :]) #turn the df entry in to an array
   # relative_density_lst_normed = np.divide(relative_density_lst, mean_nucleus) 
    
    dist_lst = np.linspace(1, len(relative_density_lst), len(relative_density_lst)) #make an array to become associated distance with each relative density entry
    dist_lst = (dist_lst - 4)# * .065 # the -4 subtraction lines up the distances with the appropriate integrated area

    name_repeated_lst = np.repeat(img_name_, len(relative_density_lst)) #this is book keeping array 
    #rand_arr = np.zeros(len(relative_density_lst))
    #rand_arr += rand

     
    df_working = pd.DataFrame({'segm_img_name' : name_repeated_lst,
                              'dist' : dist_lst,
                              'relative_density_'+category[:4] : relative_density_lst})
                              # 'mean_nucleus' : mean_lst,
                              #'relative_density_normed' : relative_density_lst_normed})
    dfs_lst_.append(df_working)
df_relative_density = pd.concat(dfs_lst_)
```
