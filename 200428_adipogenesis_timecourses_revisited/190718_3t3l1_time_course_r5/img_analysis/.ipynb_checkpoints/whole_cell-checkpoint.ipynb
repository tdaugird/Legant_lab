```python
'''generate a dictionary that has associated times for each image'''
times_4_dict = [-3, -1, 0, 0, .083, .083, .167, .167, .25, .25, .33, .33, 1, 1, 2, 2, 3, 3, 4, 4, 6, 6, -1] #list of times associated with images
times_dict = {}

for i in range(len(segm_img_name_lst)):
    img_name = segm_img_name_lst[i]
    time = times_4_dict[i]
    
    times_dict.update({img_name : time})
```

```python
times_dict
```

# Now that I have all of the organization stuff worked out it is time to actually get into the meat of this and iterate over things

```python
'''this is work horse of the script. it iterates over each of the segmented images and then findes the associated fluorescent images. it also
associates the appropriate time to the image based on dictionary above. From there it iterates over the individual objects within the
segmented images and determines, area, xy locaiton, as well as the integrated fluorescent signal within the object. The end result is 
a concatednate data frame of all of the individual cells'''

os.chdir(segm_img_dir)

df_running_lst = []


segm_img_filelist = []
for segm_img_filename in os.listdir():
    if '.tif' in segm_img_filename:
        segm_img_filelist.append(segm_img_filename)
        

for segm_img_filename in segm_img_filelist:
#for segm_img_filename in temp_lst:

    segm_img = io.imread(segm_img_filename) #read in segmented image
    
    '''the next part is finding the associated fluoresence images'''
    subdir = segm_img_filename[:-20] #generate name of subdiretory based on segmented image name
    os.chdir(orig_img_dir_base+'\\'+subdir) #change to directory with the original images
    filename_base = segm_img_filename[:-16] #generate a base filename for the fluorescent images
    
    dapi_orig_img = io.imread(filename_base+'c1.tif') #read in original dapi image
    ppar_orig_img = io.imread(filename_base+'c3.tif') #read in original ppar image
    cebp_orig_img = io.imread(filename_base+'c4.tif') #read in original cebp image
    
    labeled_img = measure.label(segm_img) #label the objects within the image 
    '''make a dataframe of object parameters'''
    df_working = pd.DataFrame(measure.regionprops_table(labeled_img, properties = ('label',  'area', 'centroid'))) #make a dataframe with a couple of shape parameters
        
    df_working['day'] = times_dict[subdir] #add appropriate day to working dataframe
    
    dapi_integrated_inten_lst = [] #lists to be populated with integrated intensities
    cebp_integrated_inten_lst = []
    ppar_integrated_inten_lst = []
    '''the next part will be iterateing over the individual objects in the segmented image and getting info regarding dapi, cebpb and pparg'''
    for nucleus in df_working.label:
        nucleus_bool_array = np.array(labeled_img == nucleus) #generate boolean mask of single cell 
        
        dapi_integrated_inten_lst.append(np.sum(np.multiply(dapi_orig_img, nucleus_bool_array)))
        cebp_integrated_inten_lst.append(np.sum(np.multiply(cebp_orig_img, nucleus_bool_array)))
        ppar_integrated_inten_lst.append(np.sum(np.multiply(ppar_orig_img, nucleus_bool_array)))
    
    '''add the integrated intensities to the working df'''
    df_working['dapi_integrated'] = dapi_integrated_inten_lst
    df_working['cebp_integrated'] = cebp_integrated_inten_lst
    df_working['ppar_integrated'] = ppar_integrated_inten_lst
    
    '''append the whole working data frame to a running list'''
    df_running_lst.append(df_working)
    
    os.chdir(segm_img_dir)
df_whole = pd.concat(df_running_lst)
```
